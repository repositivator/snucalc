```python
process_df = pd.DataFrame({'ID':list(set([int(i) for i in main_df.ID]))})

process_df["num_organ"] = 0
process_df["num_lymph"] = 0
process_df["sum_base"] = 0
process_df["sum_post"] = 0
process_df["percent_change"] = 0
process_df["lesion_size"] = "Multiple lesions"

process_df
```

```python
sum = 0
for record in range(len(process_df.index)):
    sum += 1
    
print(sum)
```

```python
sum = 0
for id in range(len(process_df.index)):

    for record in range(len(main_df.index)):
        sum += 1
        if main_df.iloc[:,0][record] == process_df.iloc[:,0][id] and re.match("lymph*", main_df.iloc[:,1][record]) != true:
            process_df["num_organ"][id] += 1
            process_df["sum_base"][id] += main_df.iloc[:,2][record] 
            process_df["sum_post"][id] += main_df.iloc[:,3][record] 
        elif main_df.iloc[:,0][record] == process_df.iloc[:,0][id] and re.match("lymph*", main_df.iloc[:,1][record]) == true:
            process_df["num_lymph"][id] += 1
            process_df["sum_base"][id] += main_df.iloc[:,2][record] 
            process_df["sum_post"][id] += main_df.iloc[:,3][record] 
print(sum)
```
